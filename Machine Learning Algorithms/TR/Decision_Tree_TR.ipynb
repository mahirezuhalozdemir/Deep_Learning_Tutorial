**Karar Ağacı**

 *   Karar ağacı, bir makine öğrenimi algoritmasıdır ve sınıflandırma ve regresyon problemleri için kullanılır. Bu algoritma, veri kümesindeki öznitelikleri kullanarak bir dizi karar kuralı oluşturur ve bu kuralları kullanarak veri noktalarını sınıflandırır veya değerlendirir.


*   Karar ağacı algoritması, ağaç yapısı kullanarak kararları temsil eder. Her iç düğüm bir özniteliği temsil eder, her dal bir öznitelik değerini ve her yaprak düğümü bir sınıf etiketi veya regresyon değerini temsil eder.

 decision.png

*   Karar ağacı algoritması, veri kümesindeki özniteliklerin değerlerine göre en iyi bölünmeyi bulmak için birçok farklı bölme kriteri kullanır.

*   Karar ağacı algoritması, modelin karmaşıklığını kontrol etmek için çeşitli yöntemler içerir, örneğin, ağacın maksimum derinliği, minimum veri noktası sayısı veya minimum bölme boyutu gibi hiperparametreler aracılığıyla. Bu, aşırı uyum (overfitting) riskini azaltmaya yardımcı olur.

*   Karar ağaçları, anlaşılması ve yorumlanması kolay olduğu için popülerdir. Ayrıca, kategorik ve sayısal özniteliklerle iyi çalışırlar ve büyük veri kümeleriyle başa çıkabilirler. Bununla birlikte, tek bir karar ağacı genellikle yüksek doğruluk sağlamaz, bu nedenle ensemble yöntemleri (örneğin, rastgele ormanlar veya gradyan artırma) genellikle daha iyi performans sağlar.

**Terminoloji**

*   **Root Node (Kök Düğüm):** Karar ağacında, karar ağacının başlangıç noktasıdır. Tüm veri setini temsil eder ve bu veri seti daha sonra homojen alt kümeler halinde bölünür.
*   **Leaf Node (Yaprak Düğüm):** Karar ağacında, bir sonuç belirtir ve daha fazla bölünemez. Yaprak düğümlerinde, sınıflandırma problemlerinde bir sınıf etiketi veya regresyon problemlerinde bir tahmin değeri bulunur.
*   **Splitting (Bölünme):** Karar ağacında, bir düğümün alt kümelerine ayırılması işlemidir. Bölünme, bir özniteliğin değerlerine göre gerçekleştirilir. Amacı, veri kümesini daha homojen alt kümeler halinde bölmektir.
*   **Branch/Sub Tree (Dal/Alt Ağaç):** Bölünmüş düğümlerden oluşan bir ağaç yapısının bir bölümüdür. Bir alt ağaç, kök düğümün alt düğümlerine ve bu düğümlerin alt düğümlerine kadar uzanır.
*   **Pruning (Budama):** Karar ağacının gereksiz dallarının veya alt ağaçlarının kaldırılması veya birleştirilmesi işlemidir. Budama, ağacın karmaşıklığını azaltır ve genelleştirme yeteneğini artırır.
*   **Parent/Child Node (Ebeveyn/Çocuk Düğüm):** Karar ağacında, bir düğümün alt düğümleri vardır. Kök düğüm, diğer düğümlerin ebeveyn düğümüdür. Diğer düğümler ise kök düğümün çocuk düğümleridir. Bu ilişki, ağacın dallarının ve alt ağaçlarının oluşumunu belirler.

**Öznitelik Seçim Ölçütleri**



**Information Gain:**
*   Bilgi kazancı (Information Gain), karar ağacı algoritmasında bölünme kriterlerini değerlendirmek için kullanılan bir ölçüdür. Bir özniteliğin bir veri kümesini ne kadar iyi böldüğünü ölçer. Bilgi kazancı yüksek olan öznitelikler, veri kümesini daha homojen alt kümeler halinde böler ve dolayısıyla daha etkili bir sınıflandırma sağlar.

*   Bilgi kazancı, bir özniteliğin bölünme öncesi ve sonrası veri kümesindeki belirsizlik düzeyindeki azalmanın ölçüsüdür. Daha az belirsizlik azalan bir bilgi kazancı sağlar ve bu da daha iyi bir bölünme olarak kabul edilir.

*   Karar ağacı algoritmasında, her özniteliğin bilgi kazancı hesaplanır ve en yüksek bilgi kazancına sahip olan öznitelik seçilerek veri kümesi bu öznitelik kullanılarak bölünür.



**Gain Ratio:**

*   Gain Ratio, karar ağacı algoritmasında kullanılan bir bölünme kriteridir. Bilgi kazancının öznitelikler arasında adil bir şekilde değerlendirilmesini sağlamak için geliştirilmiştir. Bilgi kazancı, veri kümesini ne kadar iyi böldüğünü ölçerken, Gain Ratio aynı zamanda özniteliklerin boyutlarına göre ayarlanır.

*   Gain Ratio, bir özniteliğin bilgi kazancını, özniteliğin bölünme öncesi belirsizlik düzeyine oranlar. Bu sayede büyük boyutlu özniteliklerin diğerlerine göre avantajı azaltılır. Böylece, daha fazla sayıda değere sahip olan özniteliklerin daha fazla bilgi kazanmasına engel olunur ve modelin dengesizliği azaltılır.



**Gini Index:**

*   Gini index, karar ağacı algoritmalarında sıklıkla kullanılan bir ölçüdür. Bir özniteliğin bölünme öncesi ve sonrası veri kümesindeki homojenliği veya heterojenliği ölçmek için kullanılır.

*   Gini index, bir veri kümesinin homojenliğini ölçer. Bir düğümün Gini index'i, o düğümdeki örneklerin farklı sınıflara bölünme olasılığına dayanır. Eğer bir düğümdeki tüm örnekler aynı sınıfa aitse (yani, düğüm homojen ise), Gini index 0 olur. Eğer bir düğümdeki örnekler eşit şekilde farklı sınıflara aitse (yani, düğüm tamamen heterojen ise), Gini index 0.5 olur. Dolayısıyla, Gini index'in değeri 0 ile 0.5 arasında değişir.

**Import Libraries**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

**Dataset Yükleme**

```python
dataset = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Cancer.csv')
#kanser verileri içeren dataset
```

```python
dataset.columns
```

**Özellikleri Ayırma**

```python
feature_cols = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
                'concave points_mean', 'symmetry_mean']
target_col = ['diagnosis']
X = dataset[feature_cols] # Features
y = dataset[target_col] # Target variable
```

**Verileri Ayırma**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

**Model Eğitim**

```python
model = DecisionTreeClassifier()
model = model.fit(X_train,y_train)
```

**Model Test**

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
