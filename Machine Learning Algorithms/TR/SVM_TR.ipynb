**SVM(Destek Vektör Makineleri):**
*   Eğitim verilerindeki herhangi bir noktadan en uzak olan iki sınıf arasında
karar sınırı bulan makine öğrenmesi yöntemidir.
*   Sınıflar arasında düzlem veya doğru oluşturarak sınıflandırma işlemi yapar. Bu düzlem iki sınıfın verileri için de max uzaklıkta olmalıdır.
*   Bu düzlemin belirtilen c değerinde uzaklıklarına *margin* denilir.
*   C değeri ne kadar büyük olursa margin o kadar dar olur. Overfit olma olasılığı artar.
*   Margin bölgesinde örneklerin bulunmasına 'soft margin', örnek bulunmamasına 'hard margin' denilir.
*   İki sınıflı bir SVM formülüne bakacak olursa;

w ağırlık vektörü, b sapma değeri, x inputtur.
*   Küçük ve karmaşık verisetleri ile daha iyi çalışır.
*   Hiperdüzlem ve doğru oluşturacağı için scale işlemi önemlidir.
*   2 boyut ile sınıflandırılamayan verisetleri için kernel trick uygulanır.
*   2 boyutta açıklanamayan verisetini daha fazla boyutta açıklamak için *'polynomial kernel'* kullanılır.
*   Polynomial kernel ile seçilen değişkenin değeri; overfitting durumunda azaltılır.
*   Her bir noktanın; diğer noktalara benzerliğini normal dağılımla hesaplayıp ona göre sınıflandırma işlemine *'RBF kernel'* denir.

**IMPORT İŞLEMLERİ**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

**DATASET YÜKLEME**

```python
dataset = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Fish.csv')
#7 sınıftan oluşan balık türleri veriseti
```

**DATASET AYIRMA**

```python
X = dataset.iloc[:, 2:8]
Y = dataset.iloc[:, 0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.7, random_state=42)
```

**MODEL OLUŞTURMA**

```python
model = svm.SVC(kernel='poly', degree=3, gamma='auto')
#polinom kernel trick türünde, 3 dereceli, otomatik gamma değerinde bir svm modeli oluştur.
```

**MODEL EĞİTİM**

```python
model = model.fit(X_train, Y_train)
```

**PERFORMANS DEĞERLENDİRME**

```python
result = model.predict(X_test)
cm = confusion_matrix(Y_test,result)
print('confusion matrix: \n',cm)

accuracy = accuracy_score(Y_test, result)
print('accuracy: ',accuracy)
```
