**LOGISTIC REGRESSION**
*   It uses the logistic (logit) function as the equation between input and result.
*   When we look at the logistic function equation, we see that the dependent variable y takes values between 0 and 1. (Logistic/sigmoid function)
*   Logistic regression is a classification method used to estimate the probability of a categorical dependent variable (classifications such as 0 (no), 1 (yes)...)
*   The error (cost) function is calculated more complex than linear regression (sigmoid).
*   The cost function is bounded in the range 0 and 1.

**IMPORT OPERATIONS**

```python
import numpy as np
from sklearn import linear_model
import pandas as pd
```

**DATASET UPLOAD**

```python
dataset = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Cancer.csv')
#dataset containing cancer data
```

**DEPENDENT(Y) AND INDEPENDENT(X) VARIABLE**

```python
X = np.array(dataset['radius_mean']).reshape(-1,1)  #tumour radius
Y = np.array(dataset['diagnosis'])  #tumour class(M = malignant, B = benign)
```

**LOGISTIC REGRESSION MODELLING**

```python
model = linear_model.LogisticRegression()
model.fit(X,Y)
```

**PREDICTING THE OUTPUT ACCORDING TO THE MODEL AND X VALUE**

```python
predicted = model.predict(np.array(25.46).reshape(-1,1))
print(predicted)
```
