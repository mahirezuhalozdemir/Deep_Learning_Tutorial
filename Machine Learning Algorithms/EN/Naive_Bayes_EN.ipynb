**NAIVE BAYES**
*   It is based on Bayes theorem, a formula for calculating conditional probability.
*   It is a lazy learning and generative learning  method.
*   It assumes independence between features and calculates the probability that a input class.
*   Calculates the class probability and classifies according to the biggest highest probability value.
*   The probability gives the value 0 when a input that it's not seen on train.(Zero Frequency)

**Naive Bayes Classes**

*   Gaussian Naive Bayes : The probability of features is assumed Gaussian.
*   Multinomal Naive Bayes : Used with multinomially distributed data.
*   Bernoulli Naive Bayes : Used if target value(y) is binary(0-1, yes-no)



**IMPORT OPERATIONS**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

**DATASET UPLOAD**

```python
dataset = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Diabetes.csv')
#diabet dataset
```

**DATASET SPLING**

```python
x = dataset.iloc[:, 0: 8]
#input values
y = dataset.iloc[:, 8]
#y target value
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

**BUILDING MODEL**

```python
gnb = GaussianNB()
```

**MODEL TRAINING**

```python
gnb.fit(x_train, y_train.ravel())
```

**MODEL TESTING**

```python
result = gnb.predict(x_test)
cm = confusion_matrix(y_test,result)
print('confusion matrix: \n',cm)

accuracy = accuracy_score(y_test, result)
# Sonuç : 0.96
print('accuracy: ',accuracy)
```
