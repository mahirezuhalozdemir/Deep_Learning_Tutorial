**PANDAS LIBRARY**

```python
import pandas as pd
```

**CHECK PANDAS VERSION**

```python
pd.__version__
```

**CREATE PANDAS DATAFRAME**

```python
dataset = { 'name': ['ahmet','mehmet','ali','veli'],
            'major': ['cs','medicine','math','physic']}

data_frame = pd.DataFrame(dataset)
data_frame
```

**CREATE PANDAS SERIES FROM LIST**

```python
serie = [11, 12, 13, 14, 15]
pandas_serie = pd.Series(serie)
pandas_serie
```

**CREATE INDEX LABEL**

```python
serie = [1,2,3,4]
index_serie = pd.Series(serie, index=['a','b','c','d'])
print(index_serie)
print('********')
print(index_serie['a'])
```

**PANDAS SERIES FROM DICTIONARY**

```python
dictionary = {'bmw':2014, 'audi':2020, 'porsche': 2023}
pandas_serie = pd.Series(dictionary)
pandas_serie
```
